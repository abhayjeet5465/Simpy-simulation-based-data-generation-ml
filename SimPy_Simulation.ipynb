```python
import simpy
import random
import numpy as np
```

```python
env = simpy.Environment()
```

```python
server = simpy.Resource(env, capacity=1)
```

```python
def customer(env, name, server):
    arrival_time = env.now
    print(f"{name} arrives at time {arrival_time:.2f}")

    with server.request() as request:
        yield request   # wait until server is free

        wait_time = env.now - arrival_time
        print(f"{name} starts service after waiting {wait_time:.2f}")

        service_time = random.randint(2, 5)
        yield env.timeout(service_time)

        print(f"{name} leaves at time {env.now:.2f}")
```

```python
def customer_generator(env, server):
    for i in range(5):
        env.process(customer(env, f"Customer {i+1}", server))
        inter_arrival_time = random.randint(1, 4)
        yield env.timeout(inter_arrival_time)
```

```python
env.process(customer_generator(env, server))
env.run()
```
